```python
# cheange NaN to 'Unknown'
small_df['Species '] = small_df['Species '].fillna('No Values')
#rename column
small_df = small_df.rename({'Species ':'Species'}, axis=1)
    
```

```python
Species = ['White Shark','Tiger Shark', 'Bull Shark', 'Blacktip Shark','Blue Shark', 
           'Mako Shark', 'Bronze Whaler Shark', 'Raggedtooth Shark', 'Nurse Shark',
          'Grey Nurse Shark','Zambesi Shark', 'Hammerhead Shark', 'Sandtiger Shark',
          'Basking Shark', 'Lemon Shark', 'Spinner Shark']

pattern = r"[\w]+ shark"

shark_column = []



def change_row_table(string_pattern, *table_columns):
    for line in table_columns[1:]:
        if 'shark' in line or 'Shark' in line:
            if 'involvement' in line:
                shark_column.append('Shark involvement prior to death not confirmed')
            else:
                line = ''.join(re.findall(string_pattern,line)).title()
                if line in Species:
                    shark_column.append(line)
                else: 
                    shark_column.append('Unknown')
        elif line == 'No Values':
            shark_column.append('No Values')
        else: 
            shark_column.append('Unknown')
            
change_row_table(pattern, shark_column, *small_df['Species'])
small_df['Species'] = shark_column
```
